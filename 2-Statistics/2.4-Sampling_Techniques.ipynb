## Types of Sampling Techniques

1. Probability Sampling 
2. Non-Probability Sampling

### Probability Sampling 

**1. Simple Random Sampling -**
    Every member of the population has an equal chance of being selected.
    Eg- Selecting people randomly ; Draw names randomly from a class of students

**2. Systematic Sampling -**
    Select every nth member of the population after a random starting point.
    Eg- Airport - CreditCard - 5th person,10th person, 15th person ; Feedback Survey - Select every 11th member- feedback survey

**3. Stratified Sampling -**    
    We divide the population unto strata(groups) based on specific characteristics and then randomly sampling from each strata.
    Eg- Divide employees by department and then randomly select a proportional number from each department to form a survey sample.; Age - <12, 12-18, >18

**4. Cluster Sampling -**
    Divide the population into clusters,randomly selecting clusters,then sampling all the members from the selected clusters
    Eg- Randomly selecting several school from a district and surveying all teachers within those schools

**5. Multi-Stage Sampling -**
    Combining several sampling methods usually involve selecting clusters than randomly sampling within those clusters.
    Eg- Randomly selecting cities, each selected city randomly selecting household to survey.

### Non-Probability Sampling 

Select individuals who are easiest to reach
Eg-Surveying people at mall

**1. Convenience Sampling -** 
    Selecting individuals who are easiest to reach.

**2. Judgemental (Purposive) Sampling -** 
    Select individual based on the researchers's judgement. about who would me more useful or representative.
    Eg- Choose experts in a field to participate { Data Science}

**3. Snowball Sampling -** 
    Existing study subjects recruit future subjects from among their acquaintances.
    Eg- Survey members of a rare disease
    
**4. Quota Sampling -** 
    Specific Characteristics
    Eg- Age,Gender,Group,Caste

***Selecting the sampling techniques only depends on use cases.***
