## Types of Data 
1. Quantitative (Numerical Data)
2. Qualitative (Categorical Data)

Quantiative- (Numerical Data)
1. **Discrete** - (whole numbers) eg- No. of bank accounts ; No.of children in family
2. **Continuous** - (any value) eg- Weight;Heights;Temperature;Speed

Qualitiative- (Categorical Data)
1. **Nominal** - (No Ranking) eg- Gender(M,F) ; Blood Group; Pincode
2. **Ordinal** -  (ranking is there) eg-Customer Feedback  (Good-1, Bad-2,Better-3)
