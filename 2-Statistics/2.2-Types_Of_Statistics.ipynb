## Types of Statistics 

### Descriptive Statistics

1. **Descriptive Statistics** - It involves methods for *summarizing and organizing* data to make it *understandable*.This type of statistics helps to describe features of the data in a study.

1. *Measure of Central Tendency* -[Mean,Median,Mode]
2. *Measure of Dispersion*-[Variance,Standar deviation]
3. *Data Distribution* - [Histograms,Box-plot,pie-chart,PDF,PMF]
4. *Summary Statistics* - [Five Number Summary,Q1,Q2,Q3,Maximum] 

eg- 
Mean score of students, Range of Temperatures,Histogram of ages 

### Influential Statistics

2. **Influential Statistics** - It involves methods for making predictions or inferences about a population based on a sample of data.It allows for hypothesis testing,estimation and conclusions.

( It makes conclusion from sample data about the population data.)

1. Hypothesis Testing-
2. P value
3. Confidence Interval
4. Statistical Analysis Test - [ ztest, ttest, ANOVA, ftest, Chi square ]

eg- p-value in test scores comparison, 95% confidence level for average height, Predicting house prices,Comparing test scores of different schools,Asoociation bertween gender and product prefernce 


## Example - 

Let say there are 20 statistics class in your college, and you have collected the height of students in the class.

Heights are recorded [ 175cm, 180cm, 140cm, 160cm, 120cm]

Descriptive question - "What is the average height of the entire classroom" - Masure of central tendency

Influential Question - "Are the height of the sample students in classroom similar to what you expect in the entire college ?"
