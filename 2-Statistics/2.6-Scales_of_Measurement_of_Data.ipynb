## Scales of Measurement of Data

The scales of measurement describe the nature of information within the values assigned to variables.

**4 Primary scales of Measurement-**

1. Nominal Size
2. Ordinal Scale
3. Interval
4. Ratio

### 1. Nominal Scale -
This scale classifies data into distinct catggories that do not have an intrinsic order.*(Qualitative {Categorical} Data)*

Characteristics-
- Data is characterized based on labels , names or categories
- Categories are mutually exclusive
- No logical order among categories*[No RANK]*

Examples - 
1. Gender - M;F
2. Colors - Red-5(50%) ; Pink-4(40%) ; Blue-1(10%)
3. Types of Cuisines - Italian; Chinese;Mexican


### 2. Ordinal Scale - 
This scale classifies the data into categories that can be ranked or ordered.

Characteristics-
- Data is categorized and ranked in a specific order
- The interval between ranks are not necessarily equal.

Examples- 

1. Education Level  - feature engineering can also be performed in this for ranking 
- High School - 1(Rank)
- Bachelors -   2(Rank)
- Masters -     3(Rank)
- Doctorate -   4(Rank)

2. Customer Feedback 
- Satisfied -       1(Rank)
- Very Satisfied -  2(Rank)
- Not Satisfied -   0(Rank)

3. Socio Economic Staus
- Low -     2(Rank)
- Middle -  1(Rank)
- High  -   0(Rank)

### 3. Interval Scale - 
It not only categorizes and orders but also specify the exact difference between interest.It lacks a true zero point.

Characteristics-
- Data is ordeed with consistent interval between values
- Allows for meaningfull comparison of differences[*Ratio cannot be measured*]
- No prove zero points

Examples-
1. Temperature in Fahrenheit- 
    - 10*F,20*F,30*F (Difference - 20*F-10*F=10*F; **0*F IS NOT POSSIBLE**)

2. IQ Scores
    - 90,100,110 (difference- 100-90=10; **IQ=0 IS NOT POSSIBLE**)

3. Calendar years
    - 2024,2020,2016 (difference - 2024-2020=4; **0 year IS NOT POSSIBLE**)

### 4. Ratio Scale - 
-It has all the properties of the interval scale.

Characteristics-
- The order matters
- Differences are measurable[*Ratio can be measured*]
-  Contains a 0 starting point

Example -
1. Student marks in a class-
    0,90,60,30,75,45,
    ASCENDING order-(0,30,45,60,75,90)
    (difference - 45-30=15 ; Ratio- 90/30 = 3:1)

2. Weights- 10,20,30,40(ratio - 40/10 = 4:1)

3. Income- 10,000;20,000;30,000(ratio- 30,000/10,000 = 3:1)
