## Handling Imbalanced Dataset :

- Suppose we have 1000 datapoints and in that 900 are yes and only 100 is No. Then there is biaseness towards yes. which is called imbalanced dataset.

1. **Up Sampling** - Increase data points in minority
2. **Down Sampling** - Reduce data points from majority 

```python
import numpy as np 
import pandas as pd

# Set the random seed for reproductivty 
np.random.seed(123)

#Create a dataframe with two classes 
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## Create my dataframe with inmbalanced dataset

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target':[0] *n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target':[1] *n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
## Upsampling 

df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,#Sample with replacement
         n_samples=len(df_majority),
         random_state=42
         )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```
