## File Operation - Read and Write Files 

Python provides built in fucntions and methods to read and write to files , both text and binary.This lesson will cover the basic of file handling,including - reading and writing text files and binary files 

```python
### Read a whole file 

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line 

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## .strip() removes the newline character
```

```python
## Writing a file (Overwriting)

with open('example.txt','w') as file :
    file.write('Hello world!\n')
    file.write('This is a a new line.')
```

```python
## Write a file without overwriting 
with open('example.txt','a') as file :
    file.write("Append operation taking place!\n")
```

```python
### Writing a list of lines to a file 
lines=['First line \n','Secondline \n','Third line']
with open('example.txt','a') as file:
    file.write(line)
```

```python
### Binary files 

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
### Reading a Binary files 

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from surce text file and write to a destination text file 

with open('example.txt','r') as source_file:
    content= source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
### Read a file and count the no. of lines, words and characterd.
# Counting lines, words, and characters in a text file 

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()  
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines},Words : {words},Characters :{characters}')
```

```python
### Writing and then reading a file 

#### The w+ mode in Python is used to open a file for both reading and writing.
# If the file does not exist,it will be created.If the file exists, its content is truncated(i.e., the file is overwritten)

with open('example.txt','w+') as file:
    file.write("Hello World\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
```
