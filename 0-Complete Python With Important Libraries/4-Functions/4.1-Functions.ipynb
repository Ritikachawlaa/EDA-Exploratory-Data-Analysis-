##### Introduction to Functions

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability

```python
## Syntax

def function_name(parameters):
    """Docstring"""
    # Fucntion body
    return expression
```

```python
## Why functions ?

num=24
if num%2==0:
    print("The no. is even")
else:
    print("The no. is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("The no. is even")
    else:
        print("The no. is odd")
```

```python
### Call this function
even_or_odd(24)
```

```python
### Fucntion with multiple parameters

def add(a,b):
    """This function adds two no."""
    return a+b

result=add(2,4)
print(result)
```

```python
### Default Parameters

def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet("Krish")
```

```python

def greet(name="Guest"): # Default parameter
    print(f"Hello {name} Welcome to the paradise")

greet("krish")
```

```python
## Variable length Arguments
## Positional Arguments - *args

## [instead of this we can write ] def print_numbers(num1,num2,num3,num4,num5):

def print_numbers(*krish): 
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
def print_numbers(*args): 
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
## Keywords Arguments - *kwargs

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="krish",age="32",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(val)
        print(f"Positional argument : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"krish",name="krish",age="32",country="India")
```

```python
## Return Statements

def multiply(a,b):
    return a*b
multiply(2,3)
```

```python
## Return multiple parameters 

def multiply(a,b):
    return a*b,a
multiply(2,3)
```
