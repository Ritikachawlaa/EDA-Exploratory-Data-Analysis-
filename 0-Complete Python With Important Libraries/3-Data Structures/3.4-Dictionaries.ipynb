##### Introduction to Dictionaries 

Dictionaries are unordered collection of items. They store data in key-value pairs. Keys must be unique and immutable (e.g- strings, numbers, or tuples), while values can be of any type.


```python
## Creating Dictionaries 

empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
student={"name":"krish","age":32,"name":24}
print(student)
```

```python
## accessing dictionaries elements 

student={"name":"krish","age":32,"grade":"A"}
print(student)
```

```python
## accessing dictionaries elements 

print(student['grade'])
print(student['age'])

## Accessing using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifying Dictionary Elements 
## Dictionary are mutable, so you can add, update or delete elements

print(student)
```

```python
student["age"]=33 ## Updated value for the key 
print(student)
student["address"]="India" ## Added a new key and value
print(student)
```

```python
del student['grade'] ## delete the key and value pair
print(student)
```

```python
## Dictinary methods

keys=student.keys()
print(keys)
values=student.values()
print(values)

items=student.items() ## get all key value pairs
print(items)
```

```python
## shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="krish1"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="krish2"
print(student_copy1)
print(student)
```

```python
## Iterating over dictionaries 
## you can use loops to iterate over dictionaries,(keys, values, items)

## Iterating over keys 
for keys in student.keys():
    print(keys)
```

```python
# Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs 
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries 
students={
    "student1":{"name":"krish","age":32},
    "student2":{"name":"Ritika","age":35}
}
print(students)
```

```python
## Access Nested Dictionaries Elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries 
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Examples

## Use a dictionary to count frequency of elements in list 

numbers=[1,2,2,3,3,4,4,4,4,]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into 1

dict1={"a":1,"b":2}
dict2={"c":3,"d":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
