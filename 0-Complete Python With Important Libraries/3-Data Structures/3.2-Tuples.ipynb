## Introduction To Tuples

Tuples are ordered collection of items that are immutable.They are similar to lists but their immutability makes them different.

```python
## Creatng a Tuple

empty_typle=()
print(empty_typle)
print(type(empty_typle))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
## Tuple Methods 

print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing a Tuple

packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking a tuple

a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Lists

lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(lst)
lst[0][3]
```

```python
lst[2][2]
```

```python
lst[0][0:3]
```

```python
## Nested Tuples

nested_tpl=((1,2,3),("a","b","c","d"),(True, False))
print(nested_tpl)

## Access the elements inside a tuple

print(nested_tpl[0])
print(nested_tpl[1][2])
```

```python
## Iterating over nested tuples 

for sub_tpl in nested_tpl:
    for item in sub_tpl:
        print(item,end=" ")
    print()
```
