## Operators
1. Introduction
2. Arithmetic Operatiors - Addition | Subtraction | Multiplication | Divison | Modulus | Exponentiation | Floor Divison
3. Comparison Operators - Equal to | Not equal to | Greater than | Less than | Greater than or equal to | Less than or equal to 
4. Logical Operators- AND | OR | NOT
5. Examples

```python
## Arithmetic Operations

a=10
b=5

add_result=a+b # add
sub_result=a-b # subtract
mul_result=a*b  # multiply
div_result=a/b  #division
floor_div_result=a//b  # floor division
modulus_result=a%b   # modulus 
exponent_result=a**b  # exponential
 
print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

### Logical Operators

```python
# AND
## If both true then result true

X=True
Y=True

result =X and Y
print(result)
```

```python
# OR

X=True
Y=False

result =X and Y
print(result)
```

```python
# NOT
X=False
not X
```
