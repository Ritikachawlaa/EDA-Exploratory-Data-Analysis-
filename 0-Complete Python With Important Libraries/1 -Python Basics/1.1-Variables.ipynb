```python
## Dynamic Typing
## Python allows the type of the variable to change as the program executes

var=10 #int
print(var,type(var))

var="Hello"
print(var, type(var))

var=3.14
print(var, type(var))
```

```python
## input 

age=input("What is the age")
print(age, type(age))

age=int(input("What is the age")) # typecasting
print(age, type(age))
```

```python
## Simple Calculator
num1= float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

sum = num1 + num2
difference = num1 - num2
multiplication = num1 * num2
division = num1 / num2

print("Sum is ", sum)
print("Difference is ", difference)
print("Product is ", multiplication)
print("Quotient is ", division)
```
