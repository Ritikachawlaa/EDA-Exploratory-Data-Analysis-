## Confidence Interval And Margin of Error 

C.I = 95% (2 Tail test )

**Confidence Interval** = Point Estimate +- Margin of Error 

**Eg: Question**-On the verbal section of Cat exam, The standard derivation is known to be 100.Asample of 30 test takers have a mean of 520.Construct 95% C.I about the mean.
 
**Answer** - 
lower C.I = 520 -(1.96) 100/5 = 480.8
upper C.I = 520 +(1.96) 100/5 = 559.2

**Conclusion** - I am 95% confident about the mean CAT score is between 480.8 and 559.2.
