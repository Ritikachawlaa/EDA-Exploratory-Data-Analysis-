## Types of ANOVA - 

**1. One Way ANOVA :** One factor with ateast 2 levels , and these levels are independent.

Eg- Doctor want to test a new medicine to decrease headache.
- They split the participant in 3 conditions [ 10mg, 20mg,30mg]
- Doctor ask the participants to rate the headache between 1 to 10.
Factor - Medicine 
Levels - Dosage

10mg [5,3] ;
20mg [7,4] ;
30mg [2,6]

**2. Repeated Measures ANOVA :** One factor with atleast 2 levels and levels are dependent.

Eg- Running (Factor)
    Day (Levels)
    DAY1[8,7] ;
    DAY2[5,4] ;
    DAY3[4,9]

**3. Factorial ANOVA :** Two or more factor (each of which with atleast) 2 levels, levels cn be independent and dependent.

Eg- Running , Gender (Factor)
    Day (Levels)
    Gender | Male  |Female
    DAY1[    8,9,    2,7] ;
    DAY2[    5,4,    4,8] ;
    DAY3[    4,3,    6,3]
