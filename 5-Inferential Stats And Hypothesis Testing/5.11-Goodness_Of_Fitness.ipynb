## CHI SQUARE For Goodness of Fit 

**Question**

- In 2010 census of the city, the weight of the individuals in a small city were found to be the following - 
- -     <50kg - 20%
- -     50-75 - 30%
- -     >75kg - 50%

- In 2020,weight of n=500 individuals were sampled.Below are the results- 
- -     <50kg - 140
- -     50-75 - 160
- -     >75kg - 200 
        
- Using l=0.05, would you conclude the population differences of weights has chnged in the last 10 years ?
 
- **Answer** : 
- >       Expected Data (2010)
- -          <50kg - 20%
- -          50-75 - 30%
- -          >75kg - 50%

- >       Observed Data (2020) (n=500)
- -          <50kg - 140
- -          50-75 - 160
- -          >75kg - 200 

- >       Expected Data (2020)
- -          <50kg - [0.2*500 = 100]
- -          50-75 - [0.3*500 = 150]
- -          >75kg - [0.5*500 = 250]

1. **Null Hypothesis** : H0 : The data meets the expectataion .
2. **Alternate Hypothesis** : H1 : The data does not meets the expectation.

3. **Significance value** (l=0.05)
4. **Confidence interval** (C.I) = (95%)

5. **Degree of Freedom** - df = n-1 = 3-1 = 2 (n = categories)
6. **Decision Boundary** - Critical value based on significance value needs to be find out, then test statistics(chi square test) is greater that critical value then we reject the null hypotheis.otherwise accept the null hypothesis.
7. **Critical value**  - is found out through chi -square table which is 5.991.
8. **Calculate Chi-square test statistics** - {(x^2)=Summation of([0-E]^2/E)}

- {(140-100)^2/100} + {(160-150)^2/150} + {(200-250)^2/250}
- {1600/100 + 100/150 + 2500/250 }
- {16 + 0.66 +10}
-  **26.66**

26.66>>5.99 : We reject the null Hypothesis !!

**CONCLUSION** : The weights of 2020 population are different than those accepted in 2010 popultion.
