## CHI SQUARE TEST 

- The chi square test for goodness of fit test claims about population proportions.
- It is a non-parametric test that is performed on ctegories [ ordinal and nominal] data.

**Eg**- There is a population of male who likes different color bikes. 
- **Color of Bikes**-(Yellow Bike ; Red Bike; Orange Bike )
- **Theory**- (1/3 likes red bike; 1/3 likes orange bike ; 1/3 likes yellow bike) *{ Theory Categorical distribution}*
- **Sample** - (22 - Yello Bike; 17 - Red Bike ; 59- Orange Bike) *{Observed Categorical Distribution}*

**Goodness of fit test supports the theory or not ?  - CHI SQUARE TEST**

**Question -** In a science class of 75 students, 11 are left handed .Does hthis class fit the theory that 12% of people are left handed.

- **Left handed** - [Observed - 11 ; Expected - (12/100)*75 = 9 ]
- **Right handed** - [Observed - 64 ; Expected - 66 ]
