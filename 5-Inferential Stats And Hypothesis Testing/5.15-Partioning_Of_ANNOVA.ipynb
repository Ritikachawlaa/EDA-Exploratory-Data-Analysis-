## Hypothesis Testing In ANOVA (Partitioning of Variance in the ANOVA)

1. **Null Hypothesis** - H0 : u1 = u2 = u3 ....uk
2. **Alternate Hypothesis** - H1 : Atleast one of the sample mean is not equal 
3. **F-test** :
- > Variance between Samples / Variance within Samples
- - > Variance between samples 
- - x1 : 1 2 4 5 3 (X1 = 15/5 = 3) --> Variance within Samples
- - x2 : 6 7 3 2 1 (X2 = 19/5 ) --> Variance within Samples
- - x3 : 5 6 3 2 4 (X3 = 20/5 = 4) --> Variance within Samples

### One - Way ANOVA : 
One factor with atleast 2 levels,levels are independent.

**Question**- Doctors want to test a new medication which reduces headaches.They splits the participant into 3 condition [15mg,30mg,45mg] Later on the doctor ask the patient to rate the headache between [1-10].Are there any differences between the 3 conditions alpha=0.05?

**Answer** : 
- - > 15mg : 9 8 7 8 8 9 8
- - > 30mg : 7 6 6 7 8 7 6 
- - > 45mg : 4 3 2 3 4 3 2

**1. Define Null and Alternate Hypothesis :**
- - > -  H0 : u15 =u30 = u45
- - > - H1 : not all u are equal 

**2.alpha = 0.05 ;C.I = 0.95**

**3. Degree of Freedom :** 
- - N = 21 ; a = 3 ; n = 7

- d(btn) = a-1 = 3- 1 = 2
- d(within) = N-a = 21-3 = 18
- (df1,df2) : (2,18) : used in f table with alpha value to find the critical value 
- d(total) =N-1 = 21-1 = 20 

**4.Decision Boundary :** 
- critical value - 3.5546 (If fvalue is > critical value,we reject the null hypothesis)

**5.Calculate F test Statistics :**
- - > - Sum of sqaue(SS) | - df   |- Mean square(MS)|- F value
- Between : 98.67   |   2   |   49.34   |   
- Within  : 10.29   |   18  |   0.54    |   
- Total   : 108.96  |   20  |   49.88   |   

**1. SS(btn) :**
- - > 15mg : 9+8+7+8+8+9+8 = 57
- - > 30mg : 7+6+6+7+8+7+6 = 47
- - > 45mg : 4+3+2+3+4+3+2 = 21
- - > = (57^2 + 47^2 +21^2) - (57^2 + 47^2 +21^2)/21
- - = **98.67**

**2. SS(Within) :**
- - > 15mg : 9^2+8^2+7^2+8^2+8^2+9^2+8^2 = 467 
- - > 30mg : 7^2+6^2+6^2+7^2+8^2+7^2+6^2 = 319
- - > 45mg : 4^2+3^2+2^2+3^2+4^2+3^2+2^2 = 67
- - > = 853 
- - > = 853 - [57^2+47^2+21^2]/7
- - > **= 10.29**

**3.Ftest = Ms(btn)/Ms(within)**
- - > 49.34/0.54 
- - > **= 86.56**

If f value is greater than critical value which is 3.5546 , we reject the null hypothesis : 86.56 > 3.5543 , so we reject the null hypothesis.
