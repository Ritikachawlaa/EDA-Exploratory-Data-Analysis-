## Pandas - DataFrame and Series

It is widely used for data manipulation, data analysis and data cleaning.It provides two primary data structure:Series and DataFrame.

SERIES - It is a 1-d array like object.

DATAFRAME - It is a 2-d ,size-mutable,and potentially heterogenous tabular data structure with labeled axes(rows and columns).

```python
## Series 
## Pandas Series is a 1-d array like object that can hold any data type.It is similar to a column in a table.

import pandas as pd 
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## Create a serirs from dictionary

data={'a': 1,'b':2,'c':3,'d':4}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## It is similar to a table in a database

## Create a Dataframe from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age' :[25,30,45],
    'City' :['Banglore','New York','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
### Create a Data Frame from a list of dictionaries 

data =[
   {'Name':'Krish','Age':32,'City':'Banglore'},
   {'Name':'John','Age':34,'City':'Banglore'},
   {'Name':'Bappy','Age':32,'City':'Banglore'},
   {'Name':'Jack','Age':32,'City':'Banglore'}
    ]

df=pd.DataFrame(data)
print(df)
print(type(df))
```
